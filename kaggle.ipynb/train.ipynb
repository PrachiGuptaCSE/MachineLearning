```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_csv("train.csv")
df1.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
dfle=df1
cols=['LotFrontage','MSZoning','Street','Alley','LotShape','LandContour','Utilities','LotConfig','Neighborhood','LandSlope','Condition1','Condition2','BldgType','HouseStyle','RoofStyle','RoofMatl','Exterior1st','Exterior2nd','MasVnrType','ExterQual','ExterCond','Foundation','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','Heating','HeatingQC','CentralAir','Electrical','KitchenQual','Functional','FireplaceQu','GarageType','GarageFinish','GarageQual','GarageCond','PavedDrive','PoolQC','Fence','MiscFeature','SaleType','SaleCondition']
for col in cols:
    dfle[col]=le.fit_transform(dfle[col])
    dfle
```

```python
dfle['LotFrontage']=dfle['LotFrontage'].fillna(dfle['LotFrontage'].median())
```

```python
dfle['GarageYrBlt']=dfle['GarageYrBlt'].fillna(dfle['GarageYrBlt'].median())
```

```python
dfle['MasVnrArea']=dfle['MasVnrArea'].fillna(dfle['MasVnrArea'].median())
```

```python
X=dfle.drop(['SalePrice','Id'],axis='columns')
X
```

```python
y=dfle.SalePrice
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
dfle.isnull().sum()
```

```python
null_counts=dfle.isnull().sum()
print(null_counts[null_counts>0])
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```
